```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = -np.sin(x)
y4 = np.sin(x)**2
c1,c2,c3,c4 = 'blue', 'green', 'red', 'black'
l1,l2,l3,l4 = 'sin', 'cos', '-sin', 'sin**2'
xl1, xl2, xl3, xl4 = 'x', 'x', 'x', 'x'
yl1, yl2, yl3, yl4 = 'sin', 'cos', '-sin', 'sin**2'
# create fig object
fig = plt.figure(figsize = (10,6), facecolor='lightblue')
# create subplot
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
# pass data to subplot
ax1.plot(x, y1, color=c1, label=l1)
ax2.plot(x, y2, color=c2, label=l2)
ax3.plot(x, y3, color=c3, label=l3)
ax4.plot(x, y4, color=c4, label=l4)
# create xlabel
ax1.set_xlabel(xl1)
ax2.set_xlabel(xl2)
ax3.set_xlabel(xl3)
ax4.set_xlabel(xl4)
# create ylabel
ax1.set_ylabel(yl1)
ax2.set_ylabel(yl2)
ax3.set_ylabel(yl3)
ax4.set_ylabel(yl4)
# create legend
ax1.legend(loc = 'upper right')
ax2.legend(loc = 'upper right')
ax3.legend(loc = 'upper right')
ax4.legend(loc = 'upper right')
plt.show()
```

## Plotting with sympy.plotting

The follow are sympy types of plotting
+ plot − 2D line plots

+ plot_parametric − 2D parametric plots

+ plot3d − 3D plots of functions in two variables.

+ plot3d_parametric_line - 3D line plots, defined by a parameter.

+ plot3d_parametric_surface - 3D parametric surface plots.

**plot-2D line plots**
plot(expr, range, ...)

```python
from sympy.plotting import plot # plot function is used for plotting 2d graph in sympy
from sympy import *
plot('x**2')
```

```python
from sympy.plotting import plot # plot function is used for plotting 2d graph in sympy
from sympy import *
x=symbols('x')# symbols functions returns a sympy symbol object which is stored in variable x

plot(x**2+2,line_color='red')
```

Specifying more functions in the same plot:

**plot(expr1,expr2, range)**

or

**plot((expr1, range1), (expr2, range2))**

```python
x=symbols('x')# symbols functions returns a sympy symbol object which is stored in variable x


plot((sin(x),(x, -pi, pi)),(cos(x), (x, -pi,pi)))
```

Some optional parameters in plot:
+ line_color − specifies color of the plot line.

+ title − a string to be displayed as title

+ xlabel − a string to be displayed as label for X axis

+ ylabel − a string to be displayed as label for y axis

```python
plot(sin(x),(x, 0, 4*pi),title='Sin function',legend='sin',line_color='r',
     xlabel='x',ylabel='sin(x)')
```

**Solving system of equations**

We will deal with system of equations by using numpy.

+ Unique solution:

Consider the following system of equations:
$$2x-3y+z=-1$$
$$x-y+2z=-3$$
$$3x+y-z=9$$


```python
import numpy as np
A=np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
b=np.array([-1,-3,9])
print(np.linalg.solve(A,b))
#same can be done by
#np.dot(np.linalg.inv(A),b)
```

+ No solution:

Consider the following system of equations:

$$\begin{eqnarray}
x-y+4z&=&-5\\
3x+z&=&0\\
-x+y-4z&=&20
\end{eqnarray}
$$

```python
A=np.array([[1,-1,4],[3,0,1],[-1,1,-4]])
b=np.array([-5,0,20])
np.linalg.solve(A,b)
```

```python
A=np.array([[-1,1,2],[1,2,1],[-2,-1,1]])
b=np.array([0,6,-6])

np.linalg.solve(A,b)
```

```python
######### Unique solution ##############
#A=np.array([[2,-3,1],[1,-1,2],[3,1,-1]])
#b=np.array([-1,-3,9])

######### No solution ##############
#A=np.array([[1,-1,4],[3,0,1],[-1,1,-4]])
#b=np.array([-5,0,20])
########## infinite many solutions ##########
A=np.array([[-1,1,2],[1,2,1],[-2,-1,1]])
b=np.array([0,6,-6])

Ab=np.column_stack((A,b))

if np.linalg.det(A)!=0:
    print("Solution is: ",np.linalg.solve(A,b))
elif np.linalg.matrix_rank(A)==np.linalg.matrix_rank(Ab):
    print("infinite many solutions")
else:
    print("no solution")
```
